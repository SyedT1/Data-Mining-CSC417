```python
import pandas as pd
import plotly.express as px
path = 'https://raw.githubusercontent.com/SyedT1/Data-Mining-CSC417/main/Week%201/datasets/iris.data'
cols = ['sepal_len','sepal_wid','petal_len','petal_wid','name']
dataset = pd.read_csv(path,names = cols)
print(dataset)
```

```python
species = [i for i in set(dataset['name'])]
species = species[::-1]
print(species)
```

```python
# Discarding 4th dimension
dataset = dataset.drop('petal_wid',axis=1)
```

```python

mapping = {species[i]: i for i in range(len(species))}
print(mapping['Iris-virginica'])
```

```python
dataset[dataset['name'] == species]
```

```python
# Plotting the data points in 3d plots
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
fig = plt.figure()
ax = plt.axes(projection='3d')
xAxis = dataset['sepal_len'].to_numpy()
yAxis = dataset['sepal_wid'].to_numpy()
zAxis = dataset['petal_len'].to_numpy()
X = np.array([[xAxis[0],yAxis[0],zAxis[0]]])
for i in range(1,xAxis.shape[0]):
  row = np.array([xAxis[i],yAxis[i],zAxis[i]])
  X = np.vstack((X,row))
y = np.array([])
for i in dataset['name'].astype(str).tolist():
  y = np.append(y,int(mapping[i]))
y = (np.rint(y)).astype(int)
ax = fig.add_subplot(111, projection='3d')
# Use three different shapes (triangle, square, circle) to plot data points for three different classes.
shapes = ['^', 's', 'o'] 
colors = ['r', 'g', 'b']  

for i in range(3):
    ax.scatter(X[y == i, 0], X[y == i, 1], X[y == i, 2], c=colors[i], marker=shapes[i], label=species[i])

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')


# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
class_means = []
for i in range(3):
    class_data = X[y == i]  
    class_mean = np.mean(class_data, axis=0) 
    class_means.append(class_mean)

# Group the means by class
class_labels = species
class_means_dict = dict(zip(class_labels, class_means))
for class_label, class_mean in class_means_dict.items():
    print(f"Class: {class_label}")
    print(f"Mean values: Sepal Length = {class_mean[0]}, Sepal Width = {class_mean[1]}, Petal Length = {class_mean[2]}")
```

```python
First2Dim = X[:, :2] 
First2DimNames = cols[:2]

# Now we plot histograms for the first 2 dimensions
for i in range(2):
    dimension_data = First2Dim[:, i]
    plt.hist(dimension_data, bins='auto')
    plt.xlabel(First2DimNames[i])
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {First2DimNames[i]}')
    plt.show()
```

```python
# Computing std deviation for each of the two dimensions (Sepal Length, Sepal Width).
class_stds = []
for i in range(3):
    class_data = X[y == i]  
    class_std = np.std(class_data, axis=0) 
    class_stds.append(class_std)

class_labels = species
class_std_dict = dict(zip(class_labels, class_stds))
for class_label, class_std in class_std_dict.items():
    print(f"Class: {class_label}")

    print(f"Mean values: Sepal Length = {class_std[0]}, Sepal Width = {class_std[1]}, Petal Length = {class_std[2]}")
```

```python
# drawing two separate box plot for all data points for two dimensions (Sepal Length, Sepal Width).X axis have dimensions such as sepal length or sepal width, petal length. The Y axis will have the
# box plots.
sepal_lenBxPlot = px.box(title="Sepal Length",data_frame = dataset,y = cols[0],points = "all")
sepal_lenBxPlot.show()

sepal_widBxPlot = px.box(title = "Sepal Width",data_frame = dataset,y = cols[1],points = "all")
sepal_widBxPlot.show()

```
